```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
df.info()
df.describe(include ="all")
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

---------------- 1.COUNT MOVIES VS TV SHOWS ---------------------------

```python
df["type"].value_counts()
```

```python
sns.countplot(data= df, x="type")
plt.title("MOVIES VS TV SHOWS")
plt.show()
```

---------------------- 2.Top 10 genres -------------------

```python
df["listed_in"] = df["listed_in"].astype(str)
genres = df["listed_in"].str.split(",",expand=True).stack()
genres.value_counts().head(10)
```

```python
genres.value_counts().head(10).plot(kind="bar")
plt.title("TOP 10 GENRES ON NETFLIX")
plt.xlabel("genres")
plt.ylabel("count")
plt.show()
```

-------------------- 3.No.of releases per year ----------------------

```python
df['release_year'].value_counts().sort_index().plot(kind='line')
plt.title("content release trend over year")
plt.xlabel("year");plt.ylabel("no.of titles")
plt.show()
```

-------------- 4.rating distribution -------------------

```python
df['rating'].value_counts().plot(kind='bar')
plt.title("Distribution of Ratings")
plt.xlabel("Rating"); plt.ylabel("Count")
plt.show()
```

------ 5.Which countries produce the most Netflix content? --------------

```python
df['country'] = df['country'].astype(str)
countries = df['country'].str.split(', ', expand=True).stack()
countries.value_counts().head(10)
```

```python
countries.value_counts().head(10).plot(kind='bar')
plt.title("Top Countries Producing Netflix Content")
plt.show()
```
